```python
"""
COMP2152 - Assignment 1
Lab Professor: Laily Ajellu
@author: Israr Wahid - 101348701
"""

import pickle

employee_list = []
item_list = []

# This function registers an employee into the system with the necessary information
def create_employee():
    while True:
        while True:
            str_input = input("Enter employee ID: ")
            try:
                employee_id = int(str_input)
                employee_exist = 0
                for i in range(0, len(employee_list)):
                    if employee_id in employee_list[i]:
                        employee_exist = 1
                        print(employee_id, " already exist")
                if employee_exist == 0:
                    break
            except ValueError:
                print("Employee ID must be a number")
        name = input("Enter employee name: ")
        while True:
            employee_type = input("Enter employee type: ")
            if not (employee_type == "hourly" or employee_type == "manager"):
                print("Employee type must be either hourly or manager")
            else:
                break
        while True:
            year = input("Enter years worked: ")
            try:
                temp = int(year)
                year = temp
                break
            except ValueError:
                print("Years worked must be a number")
        while True:
            str_input = input("Enter discount number: ")
            try:
                discount_number = int(str_input)
                discount_number_exists = 0
                for i in range(0, len(employee_list)):
                    if discount_number in employee_list[i]:
                        discount_number_exists = 1
                        print(discount_number, " already exist")
                if discount_number_exists == 0:
                    break
            except ValueError:
                print("Employee discount number must be a number")
        info_list = [employee_id, name, employee_type, year, 0, 0, discount_number]
        employee_list.append(info_list)

        save_data()

        n = input("Do you want to enter another employee (N/n for no): ")
        if n == 'N' or n == 'n':
            break


# This function registers an item into the system with the necessary information
def create_item():
    while True:
        while True:
            str_input = input("Enter an item number: ")
            try:
                item_number = int(str_input)
                item_number_exists = 0
                for i in range(0, len(item_list)):
                    if item_number in item_list[i]:
                        item_number_exists = 1
                        print(item_number, " already exist")
                if item_number_exists == 0:
                    break
            except ValueError:
                print("Item number must be a number")
        name = input("Enter item name: ")
        while True:
            str_input = input("Enter item cost: ")
            try:
                cost = int(str_input)
                break
            except ValueError:
                print("Item cost must be a number")
        item_info_list = [item_number, name, cost]
        item_list.append(item_info_list)

        save_data()

        n = input("Do you want to enter another item (N/n for no): ")
        if n == 'N' or n == 'n':
            break


# This function allows the user to make the purchase with the discount amount based on the numbers of years worked
def make_purchase():
    print("Item number | Item Name       | Item Cost")
    for i in range(0, len(item_list)):
        print("%11d" % item_list[i][0], "| ", "%14s" % item_list[i][1], "| $", item_list[i][2])
    while True:
        while True:
            employee_discount_number = int(input("Enter employee discount number: "))
            discount_number_exists = 0
            employee_index = 0
            for i in range(0, len(employee_list)):
                if employee_discount_number in employee_list[i]:
                    discount_number_exists = 1
                    employee_index = i
                    break
            if discount_number_exists == 1:
                break
            else:
                print(employee_discount_number, " employee discount number does not exist")
        while True:
            num = int(input("Enter item number: "))
            item_number_exists = 0
            item_index = 0
            for i in range(0, len(item_list)):
                if num in item_list[i]:
                    item_number_exists = 1
                    item_index = i
                    break
            if item_number_exists == 1:
                break
            else:
                print(num, " item does not exist")
        confirm = input("Do you want to make this purchase (Y/y for yes): ")
        if confirm == 'Y' or confirm == 'y':
            cost = item_list[item_index][2]
            discount = 0
            if employee_list[employee_index][5] < 200:
                discount = int(employee_list[employee_index][3]) * 2
                if discount > 10:
                    discount = 10
                if employee_list[employee_index][2] == "manager":
                    discount += 10
            cost = cost - ((cost * discount) / 100)
            employee_list[employee_index][4] += cost
            employee_list[employee_index][5] += discount

            save_data()
        confirm = input("Do you want to make a new purchase (N/n for no): ")
        if confirm == 'N' or confirm == 'n':
            break
    employee_summary()


# This function provides the information for all the employees who are registered in the company
def employee_summary():
    print(
        "Employee ID | Employee Name | Employee Type | Years Worked | Total Purchased | Total Discount | Employee "
        "Discount Number")
    for i in range(0, len(employee_list)):
        total_purchase = "${:,.2f}".format(float(employee_list[i][4]))
        total_discount = "${:,.2f}".format(float(employee_list[i][5]))
        print("%11d" % employee_list[i][0], "| ", "%12s" % employee_list[i][1], "| ", "%12s" % employee_list[i][2],
              "| ", "%11d" % employee_list[i][3], "| ", "%14s" % total_purchase, "| ", "%13s" % total_discount, "| ",
              employee_list[i][6])


# This function saves the data in file so there is no need to re-enter each time new data to test the program
def save_data():
    with open('item_list.txt', 'wb') as fp:
        pickle.dump(item_list, fp)

    with open('employee_list.txt', 'wb') as fp:
        pickle.dump(employee_list, fp)


# This function allows opening and reading the saved data,
# which saves time from entering new data each time to test the program
def read_data():
    global item_list
    global employee_list

    try:
        with open('item_list.txt', 'rb') as fp:
            item_list = pickle.load(fp)

        with open('employee_list.txt', 'rb') as fp:
            employee_list = pickle.load(fp)

    except FileNotFoundError:
        item_list = []
        employee_list = []


# This is the main entry point from where the execution starts
if __name__ == "__main__":
    read_data()

    while True:
        print("-------------------------------------------------")
        print("| 1- Create Employee                            |")
        print("| 2- Create Item                                |")
        print("| 3- Make Purchase                              |")
        print("| 4- All Employee Summary                       |")
        print("| 5- Exit                                       |")
        print("-------------------------------------------------")
        menu_choice = int(input("Enter your choice by entering the number beside each action: "))
        if menu_choice == 1:
            create_employee()
        elif menu_choice == 2:
            create_item()
        elif menu_choice == 3:
            make_purchase()
        elif menu_choice == 4:
            employee_summary()
        elif menu_choice == 5:
            break




```
